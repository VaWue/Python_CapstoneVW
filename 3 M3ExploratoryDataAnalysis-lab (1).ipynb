## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
# Plot a vertical box plot for ConvertedComp using Pandas
plt.figure(figsize=(8, 6))
df['ConvertedComp'].plot(kind='box', patch_artist=True)
plt.title('Box Plot of ConvertedComp')
plt.ylabel('Converted Compensation (USD)')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
# Calculate Interquartile Range (IQR) for ConvertedComp
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

print(f"The Interquartile Range (IQR) for ConvertedComp is: ${IQR:.2f}")
```

Find out the upper and lower bounds.


```python
# your code goes here
# Calculate Upper and Lower Bounds for ConvertedComp
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

print(f"The Upper Bound for ConvertedComp is: ${upper_bound:.2f}")
print(f"The Lower Bound for ConvertedComp is: ${lower_bound:.2f}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
# Identify outliers in ConvertedComp
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]

# Count the number of outliers
num_outliers = outliers.shape[0]

print(f"The number of outliers in the ConvertedComp column is: {num_outliers}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# Create a new DataFrame without outliers in ConvertedComp
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Display the shape of the new DataFrame
print(f"Shape of DataFrame without outliers: {df_no_outliers.shape}")

# Optionally, display the first few rows of the new DataFrame
# print(df_no_outliers.head())
```
